```python
#텍스트 전처리
#한글 제외문자 제거
import re

remove_except_ko=re.compile(r"[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]")
def preprocess(text):
    text=re.sub(remove_except_ko, ' ', text).strip()
    return text
```

```python
content=preprocess(content)
```

```python
content
```

```python
#mecab 토큰화
from konlpy.tag import Mecab
#선언
mecab=Mecab()
mecab_tokens = mecab.morphs(content)
print(mecab_tokens[:30])
```

```python
mecab_tokens2=mecab_tokens.copy()
```

```python
mecab_tokens2.count('코로나')
```

```python
mecab_tokens2.count('배달')
```

```python
#mecab pos-tagging
mecabTag=[]
for token in mecab_tokens:
    mecabTag +=mecab.pos(token)

print(mecabTag[:30])
```

```python
#mecab
#최빈어 조회, 최빈어=불용어로 제거

from collections import Counter
Counter(mecabTag).most_common()
```

```python
#불용어 처리
stopPos = ['JKS','JKC','JKG','JKO','JKB','JKV','JKQ','JC','JX',
           'SF','SE','SS','SP','SSO','SSC','SC','SY','SH','SL','SN',
           'MM','MAG','MAJ']
stopWord = ['오류', '우회', '하','을','다','는', '저작권자', '무단', '전재', '재', '배포', '금지', '송고']

tokenized_text=[]
for tag in mecabTag:
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
            tokenized_text.append(tag[0])

print(tokenized_text)
```

```python
# 감성분석을 통한 톤 분석
```

```python
# 사전 정의
positive_dictionary = pd.read_csv("C:/Users/lg/Desktop/00study/KnuSentiLex-master/positive.csv")
negative_dictionary = pd.read_csv("C:/Users/lg/Desktop/00study/KnuSentiLex-master/negative.csv")
```

```python
#사전 리스트 전환
pos_list=list(positive_dictionary['word'])
neg_list=list(negative_dictionary['word'])
```

```python
pos = 0
for token in tokenized_text:
    if token in pos_list:
        pos+=1
print("긍정 :", pos)
```

```python
posCount=[]
posCount.append(pos)
```

```python
neg=0
for token in tokenized_text:
    if token in neg_list:
        neg+=1
print("부정:", neg)
```

```python
negCount=[]
negCount.append(neg)
```

```python
#문장 어조 구하기
tone = (pos-neg)/(pos+neg)
tone
tone_list=[]
tone_list.append(tone)
print(tone_list)
```
